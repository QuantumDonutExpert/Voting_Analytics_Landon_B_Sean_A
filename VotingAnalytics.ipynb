```python
import pandas as pd
import matplotlib.pyplot as plt

races = {
    'White': {'male': white_alone['male'], 'female': white_alone['female']},
    'Black': {'male': black_alone['male'], 'female': black_alone['female']},
    'Asian': {'male': asian_alone['male'], 'female': asian_alone['female']},
    'Hispanic/Latino': {'male': hispanic_or_latino_alone['male'], 'female': hispanic_or_latino_alone['female']}
}

race_names = list(races.keys())
bar_width = 0.4

fig, ax = plt.subplots(figsize=(12, 6))

for i, race in enumerate(race_names):
    male_voted = races[race]['male']['Reported Voted (Percent)'].iloc[0] if 'Reported Voted (Percent)' in races[race]['male'].columns else 0
    female_voted = races[race]['female']['Reported Voted (Percent)'].iloc[0] if 'Reported Voted (Percent)' in races[race]['female'].columns else 0

    ax.bar(i - bar_width/2, male_voted, width=bar_width, label='Male' if i == 0 else '', color='navy')
    ax.bar(i + bar_width/2, female_voted, width=bar_width, label='Female' if i == 0 else '', color='pink')

    ax.text(i - bar_width/2, male_voted + 1, f'{male_voted:.1f}%', ha='center', va='bottom', fontsize=10)
    ax.text(i + bar_width/2, female_voted + 1, f'{female_voted:.1f}%', ha='center', va='bottom', fontsize=10)

ax.set_xlabel("Race", fontsize=12)
ax.set_ylabel("Voted Percentage", fontsize=12)
ax.set_title("Voting Percentages by Race and Gender", fontsize=14)
ax.set_xticks(range(len(race_names)))
ax.set_xticklabels(race_names)
ax.set_ylim(0, 100)
ax.legend()

plt.show()
```

```python
import plotly.graph_objects as go
import us

state_abbreviations_alphabetical = [state.abbr for state in us.states.STATES]
state_abbreviations_alphabetical.sort()

state_names = list(states_info.keys())
state_names.remove('us')

heatmap_state_abbreviations = []

for state_name in state_names:
    try:
        abbreviation = next(abbr for abbr in state_abbreviations_alphabetical
                            if us.states.lookup(abbr).name == state_name)

        heatmap_state_abbreviations.append(abbreviation)
    except StopIteration:
        print(f"Warning: Could not find abbreviation for {state_name}, skipping.")


vote_percentages = []
for state in state_names:
    vote_percentages.append(states_info[state]['Percent Voted (Total)'].iloc[0])

fig = go.Figure(data=go.Choropleth(
    locations=heatmap_state_abbreviations,
    z=vote_percentages,
    locationmode='USA-states',
    colorscale='Blues',
    colorbar_title="Voting Percentage",
))

fig.update_layout(
    title_text='US Voting Percentage by State',
    geo_scope='usa',
)

fig.show()
```

```python
import plotly.graph_objects as go
import us



def create_race_voting_map(race):
    state_abbreviations = []
    vote_percentages = []

    for state_name in states_info:
        if state_name != 'us':
            try:
                row_index = states_info[state_name][
                    states_info[state_name]['Sex, Race, and Hispanic Origin'].str.contains(race)
                ].index[0]

                state_abbreviation = us.states.lookup(state_name).abbr
                voting_percentage = states_info[state_name]['Percent Voted (Total)'].iloc[row_index]

                state_abbreviations.append(state_abbreviation)
                vote_percentages.append(voting_percentage)

            except (IndexError, AttributeError):
                print(f"Skipping {state_name} due to missing race data")
                continue
            except Exception as e:
                print(f"An error occurred: {e}")


    fig = go.Figure(data=go.Choropleth(
        locations=state_abbreviations,
        z=vote_percentages,
        locationmode='USA-states',
        colorscale='Blues',
        colorbar_title="Voting Percentage",
    ))

    fig.update_layout(
        title_text=f'US Voting Percentage by State ({race})',
        geo_scope='usa',
    )
    return fig


races = ['White alone', 'Black or African American alone', 'Asian alone', 'Hispanic or Latino']

for race in races:
    fig = create_race_voting_map(race)
    fig.show()
    # Optional: Save each map to a file
    # fig.write_image(f"/content/drive/MyDrive/voting_map_{race}.png")
```

```python
import plotly.graph_objects as go
import plotly.io as pio
import us

state_abbreviations_alphabetical = [state.abbr for state in us.states.STATES]
state_abbreviations_alphabetical.sort()

state_names = list(states_info_age.keys())
state_names.remove('us')

heatmap_state_abbreviations = []
for state_name in state_names:
    try:
        abbreviation = next(abbr for abbr in state_abbreviations_alphabetical
                            if us.states.lookup(abbr).name == state_name)
        heatmap_state_abbreviations.append(abbreviation)
    except StopIteration:
        print(f"Warning: Could not find abbreviation for {state_name}, skipping.")

def get_age_vote_percentage(state_data, age_group):
    try:
        row_index = state_data[state_data.iloc[:, 0] == age_group].index[0]
        return state_data['Percent Voted (Total)'].iloc[row_index]
    except IndexError:
        return 0

age_groups = states_info_age['us']['Age'].tolist()

for age_group in age_groups:
    vote_percentages = []
    for state in state_names:
        vote_percentages.append(get_age_vote_percentage(states_info_age[state], age_group))

    fig = go.Figure(data=go.Choropleth(
        locations=heatmap_state_abbreviations,
        z=vote_percentages,
        locationmode='USA-states',
        colorscale='Blues',
        colorbar_title=f"Voting Percentage ({age_group})",
    ))

    fig.update_layout(
        title_text=f'US Voting Percentage by State ({age_group})',
        geo_scope='usa',
    )

    #file_path = f'/content/drive/MyDrive/voting_percentage_heatmap_{age_group}.png'
    #pio.write_image(fig, file_path)

    fig.show()
```
