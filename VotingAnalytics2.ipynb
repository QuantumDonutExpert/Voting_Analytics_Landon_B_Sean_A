```python
temp_df = pd.read_excel('/content/drive/MyDrive/all_races.xlsx', header=6, nrows=1)
num_cols = len(temp_df.columns)

all_races = {
    "both_sexes": pd.read_excel('/content/drive/MyDrive/all_races.xlsx', header=6, skiprows=0, usecols=range(1, num_cols), nrows=6),
    "male": pd.read_excel('/content/drive/MyDrive/all_races.xlsx', header=6, skiprows=6, usecols=range(1, num_cols), nrows=6),
    "female": pd.read_excel('/content/drive/MyDrive/all_races.xlsx', header=6, skiprows=12, usecols=range(1, num_cols), nrows=6)
}

new_column_names = ['Age', 'Total Population', 'Total Citizen Population', 'RR (Number)',
                    'RR (Percent)', 'RnR (Number)', 'RnR (Percent)', 'No Response (Number)',
                    'No Response (Percent)', 'RV (Number)', 'RV (Percent)', 'RnV (Number)',
                    'RnV (Percent)', 'NRV (Number)', 'NRV (Percent)',
                    'Reported Registered (Percent)', 'Reported Voted (Percent)']

for i in all_races:
    old_column_names = all_races[i].columns
    column_mapping = dict(zip(old_column_names, new_column_names))
    all_races[i] = all_races[i].rename(columns=column_mapping)

print(all_races["female"])
```

```python
import matplotlib.pyplot as plt

races = {
    'White': {'male': white_alone['male'], 'female': white_alone['female']},
    'Black': {'male': black_alone['male'], 'female': black_alone['female']},
    'Asian': {'male': asian_alone['male'], 'female': asian_alone['female']},
    'Hispanic/Latino': {'male': hispanic_or_latino_alone['male'], 'female': hispanic_or_latino_alone['female']}
}

plt.figure(figsize=(12, 6))

x = 0
width = 0.35

for gender, color in zip(['male', 'female'], ['skyblue', 'lightcoral']):
  voted_percentages = [races[race][gender].iloc[0, 16] for race in race_names]
  plt.bar([i + x for i in range(len(race_names))], voted_percentages, width, label=gender.capitalize(), color=color)
  for i, percentage in enumerate(voted_percentages):
    plt.text(i + x, percentage + 1, f'{percentage:.1f}%', ha='center', va='bottom', fontsize=8)
  x += width

plt.xlabel("Race", fontsize=12)
plt.ylabel("Voted Percentage", fontsize=12)
plt.title("Voting Percentages by Race and Gender", fontsize=14)
plt.xticks([i + width / 2 for i in range(len(race_names))], race_names)
plt.ylim(0, 100)
plt.legend()
plt.tight_layout()
#plt.savefig('/content/drive/My Drive/B.png')
plt.show()
```

```python
import matplotlib.pyplot as plt

races_data = {
    'White': {'male': white_alone['male'], 'female': white_alone['female']},
    'Black': {'male': black_alone['male'], 'female': black_alone['female']},
    'Asian': {'male': asian_alone['male'], 'female': asian_alone['female']},
    'Hispanic/Latino': {'male': hispanic_or_latino_alone['male'], 'female': hispanic_or_latino_alone['female']}
}

plt.figure(figsize=(12, 6))

x = 0
width = 0.35

handles = []
labels = []

for race, data in races_data.items():
    male_reg = data['male'].iloc[0, 15]
    female_reg = data['female'].iloc[0, 15]

    handle_male = plt.bar(x - width/2, male_reg, width, color='skyblue')
    handle_female = plt.bar(x + width/2, female_reg, width, color='lightcoral')

    if x == 0:
        handles.extend([handle_male, handle_female])
        labels.extend(['Male', 'Female'])

    plt.text(x - width/2, male_reg + 1, f'{male_reg:.1f}%', ha='center', va='bottom', fontsize=8)
    plt.text(x + width/2, female_reg + 1, f'{female_reg:.1f}%', ha='center', va='bottom', fontsize=8)
    x += 1

plt.xlabel("Race", fontsize=12)
plt.ylabel("Registration Percentage", fontsize=12)
plt.title("Registration Percentages by Race and Sex", fontsize=14)
plt.xticks(range(len(races_data)), races_data.keys())
plt.ylim(0, 100)

plt.legend(handles, labels)

plt.tight_layout()
#plt.savefig('/content/drive/My Drive/C.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

races = {
    'White': white_alone['both_sexes'],
    'Black': black_alone['both_sexes'],
    'Asian': asian_alone['both_sexes'],
    'Hispanic/Latino': hispanic_or_latino_alone['both_sexes']
}

age_brackets = ['18+', '18-24', '25-44', '45-64', '65-74', '75+']
colors = ['blue', 'orange', 'green', 'red']
plt.figure(figsize=(12, 8))

for i, (race_name, race_data) in enumerate(races.items()):
    registration_percentages = [race_data.iloc[j, 15] for j in range(6)]
    plt.plot(age_brackets, registration_percentages, marker='o', color=colors[i], label=race_name)

plt.xlabel("Age Brackets", fontsize=14)
plt.ylabel("Registration Percentage", fontsize=14)
plt.title("Registration Percentage by Age and Race", fontsize=16)
plt.legend(fontsize=12)
plt.grid(True)
plt.ylim(0, 100)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
#plt.savefig('/content/drive/My Drive/E.png')
plt.show()
```
